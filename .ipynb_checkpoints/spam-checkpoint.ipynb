```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./emails.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data_X = data.drop(columns=['Prediction', 'Email No.'])
data_Y = data.Prediction
```

```python

data_Y.head(1)
```

```python
data_X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
input_X = scaler.fit_transform(data_X)
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(input_X, data_Y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression( max_iter=350,C=1)
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy*100)
```

```python
from xgboost import XGBClassifier
xbg_model = XGBClassifier(booster='gblinear', max_depth = 4)
xbg_model.fit(
    X_train, Y_train,
    eval_set = [(x_test, y_test)],
    verbose = 1,
    early_stopping_rounds = 10
)
```

```python
xbg_model.score(X_train, Y_train)
```

```python
xbg_model.score(x_test, y_test)
```
